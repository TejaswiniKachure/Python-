```python
# Challenge: Student Marks Analyzer

# Create a Python program that allows a user to input student names along with their marks and then calculates useful statistics.
# Your program should:
# 1. Let the user input multiple students with their marks (name + integer score).
# 2. After input is complete, display:
#    - Average marks
#    - Highest marks and student(s) who scored it
#    - Lowest marks and student(s) who scored it
#    - Total number of students

# Bonus:
# - Allow the user to enter all data first, then view the report
# - Format output clearly in a report-style layout
# - Prevent duplicate student names

def student_marks_analyzer():
    students = {}

    print("=== Student Marks Entry ===")

    while True:
        name = input("Enter student name (or type 'done' to finish): ").strip()

        if name.lower() == "done":
            break

        # Prevent duplicate names
        if name in students:
            print("⚠️ Student already exists! Try a different name.")
            continue

        try:
            marks = int(input("Enter marks: "))
            students[name] = marks
        except ValueError:
            print("❌ Invalid marks! Please enter an integer.")

    if not students:
        print("\nNo data entered.")
        return

    # ------------------ Analysis ------------------
    total_students = len(students)
    total_marks = sum(students.values())
    average = total_marks / total_students

    max_marks = max(students.values())
    min_marks = min(students.values())

    top_students = [name for name, m in students.items() if m == max_marks]
    low_students = [name for name, m in students.items() if m == min_marks]

    # ------------------ Report ------------------
    print("\n" + "="*40)
    print("📊 STUDENT MARKS REPORT")
    print("="*40)

    print(f"Total Students : {total_students}")
    print(f"Average Marks  : {average:.2f}")

    print(f"\nHighest Marks  : {max_marks}")
    print("Topper(s)      : " + ", ".join(top_students))

    print(f"\nLowest Marks   : {min_marks}")
    print("Lowest Scorer(s): " + ", ".join(low_students))

    print("="*40)


# Run program
student_marks_analyzer()
```

```python
# Challenge: CLI Contact Book (CSV-Powered)

# Create a terminal-based contact book tool that stores and manages contacts using a CSV file.

# Your program should:
# 1. Ask the user to choose one of the following options:
#    - Add a new contact
#    - View all contacts
#    - Search for a contact by name
#    - Exit
# 2. Store contacts in a file called `contacts.csv` with columns:
#    - Name
#    - Phone
#    - Email
# 3. If the file doesn't exist, create it automatically.
# 4. Keep the interface clean and clear.

# Example:
# Add Contact
# View All Contacts
# Search Contact
# Exit

# Bonus:
# - Format the contact list in a table-like view
# - Allow partial match search
# - Prevent duplicate names from being added


import csv
import os

FILE_NAME = "contacts.csv"


# ------------------ Initialize File ------------------
def initialize_file():
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["Name", "Phone", "Email"])


# ------------------ Load Contacts ------------------
def load_contacts():
    contacts = []
    with open(FILE_NAME, "r") as f:
        reader = csv.DictReader(f)
        for row in reader:
            contacts.append(row)
    return contacts


# ------------------ Save Contacts ------------------
def save_contacts(contacts):
    with open(FILE_NAME, "w", newline="") as f:
        fieldnames = ["Name", "Phone", "Email"]
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(contacts)


# ------------------ Add Contact ------------------
def add_contact():
    contacts = load_contacts()

    name = input("Enter Name: ").strip()

    # Duplicate check
    if any(c["Name"].lower() == name.lower() for c in contacts):
        print("⚠️ Contact already exists!")
        return

    phone = input("Enter Phone: ").strip()
    email = input("Enter Email: ").strip()

    contacts.append({"Name": name, "Phone": phone, "Email": email})
    save_contacts(contacts)

    print("✅ Contact added successfully!")


# ------------------ View Contacts ------------------
def view_contacts():
    contacts = load_contacts()

    if not contacts:
        print("No contacts found.")
        return

    print("\n📒 CONTACT LIST")
    print("-" * 50)
    print(f"{'Name':<15}{'Phone':<15}{'Email'}")
    print("-" * 50)

    for c in contacts:
        print(f"{c['Name']:<15}{c['Phone']:<15}{c['Email']}")

    print("-" * 50)


# ------------------ Search Contact ------------------
def search_contact():
    contacts = load_contacts()
    keyword = input("Enter name to search: ").lower()

    results = [c for c in contacts if keyword in c["Name"].lower()]

    if not results:
        print("No matching contact found.")
        return

    print("\n🔍 SEARCH RESULTS")
    print("-" * 50)
    for c in results:
        print(f"{c['Name']:<15}{c['Phone']:<15}{c['Email']}")
    print("-" * 50)


# ------------------ Edit Contact ------------------
def edit_contact():
    contacts = load_contacts()
    name = input("Enter contact name to edit: ").lower()

    for c in contacts:
        if c["Name"].lower() == name:
            print("Leave blank to keep old value")

            new_phone = input(f"New Phone ({c['Phone']}): ").strip()
            new_email = input(f"New Email ({c['Email']}): ").strip()

            if new_phone:
                c["Phone"] = new_phone
            if new_email:
                c["Email"] = new_email

            save_contacts(contacts)
            print("✅ Contact updated!")
            return

    print("Contact not found.")


# ------------------ Delete Contact ------------------
def delete_contact():
    contacts = load_contacts()
    name = input("Enter contact name to delete: ").lower()

    new_contacts = [c for c in contacts if c["Name"].lower() != name]

    if len(new_contacts) == len(contacts):
        print("Contact not found.")
    else:
        save_contacts(new_contacts)
        print("🗑️ Contact deleted successfully!")


# ------------------ Main Menu ------------------
def main():
    initialize_file()

    while True:
        print("\n===== 📞 CONTACT BOOK =====")
        print("1. Add Contact")
        print("2. View Contacts")
        print("3. Search Contact")
        print("4. Edit Contact")
        print("5. Delete Contact")
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_contact()
        elif choice == "2":
            view_contacts()
        elif choice == "3":
            search_contact()
        elif choice == "4":
            edit_contact()
        elif choice == "5":
            delete_contact()
        elif choice == "6":
            print("Exiting... 👋")
            break
        else:
            print("❌ Invalid choice!")


# ------------------ Run ------------------
if __name__ == "__main__":
    main()
```

```python
#  Challenge:  Personal Movie Tracker with JSON

# Create a Python CLI tool that lets users maintain their own personal movie database, like a mini IMDb.

# Your program should:
# 1. Store all movie data in a `movies.json` file.
# 2. Each movie should have:
#    - Title
#    - Genre
#    - Rating (out of 10)
# 3. Allow the user to:
#    - Add a movie
#    - View all movies
#    - Search movies by title or genre
#    - Exit the app

# Bonus:
# - Prevent duplicate titles from being added
# - Format output in a clean table
# - Use JSON for reading/writing structured data


import json
import os

file_name = "movies.json"

# Create file if not exists
if not os.path.exists(file_name):
    with open(file_name, "w") as f:
        json.dump([], f)

while True:
    print("\n--- Movie Tracker ---")
    print("1. Add Movie")
    print("2. View Movies")
    print("3. Search Movie")
    print("4. Exit")

    choice = input("Enter choice: ")

    #Add Movie
    if choice == "1":
        title = input("Enter title: ")
        genre = input("Enter genre: ")
        rating = input("Enter rating (0-10): ")

        with open(file_name, "r") as f:
            movies = json.load(f)

        # Check duplicate
        found = False
        for m in movies:
            if m["title"].lower() == title.lower():
                found = True
                break

        if found:
            print("Movie already exists!")
        else:
            movies.append({
                "title": title,
                "genre": genre,
                "rating": rating
            })

            with open(file_name, "w") as f:
                json.dump(movies, f)

            print("Movie added!")

    #View Movies
    elif choice == "2":
        with open(file_name, "r") as f:
            movies = json.load(f)

        if len(movies) == 0:
            print("No movies found.")
        else:
            print("\nMovie List:")
            for m in movies:
                print(m["title"], "-", m["genre"], "-", m["rating"])

    #Search Movie
    elif choice == "3":
        keyword = input("Enter name or genre: ").lower()

        with open(file_name, "r") as f:
            movies = json.load(f)

        found = False
        for m in movies:
            if keyword in m["title"].lower() or keyword in m["genre"].lower():
                print(m["title"], "-", m["genre"], "-", m["rating"])
                found = True

        if not found:
            print("No movie found.")

    #Exit
    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice!")
```
